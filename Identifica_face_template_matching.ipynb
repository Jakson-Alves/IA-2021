# **Template matching - identificando Messi**


```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
!unzip imagens.zip -d .
```

```python
img_rgb = cv2.imread('messi.png')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY) #Converte a imagem

#O 0 no final pega com escala de cinza para não precisar converter como acima
template = cv2.imread('messi_face.jpg', 0)

#Pega o tamanho da imagem
w, h = template.shape
print(w, h)

# Res cria uma matriz a partir do tamplate
res = cv2.matchTemplate(img_gray, template, cv2.TM_CCOEFF_NORMED)

# Localiza os valores mínimos e máximos (onde está a face)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
# print(min_val, max_val, min_loc, max_loc)

# Pega a localização do ponto, pega a largura e altura do template para fazer o ponto
bottom_right = (max_loc[0] + w, max_loc[ 1] + w)

#Cria o retangulo ao redor da face do messi, com a cor verde e expessura 2
cv2.rectangle(img_rgb, max_loc, bottom_right, ( 0, 255, 0), 2)

#Mostra a imagem
cv2_imshow(img_rgb)
```
