# **Identifica se é Banana ou Laranja através do pH e Peso**

```python
import numpy as np

#Valor de PH e peso da laranja e banana
pH = np.array([6.8, 4.7, 6, 3, 2.9, 4.5])
peso = np.array([113, 122, 107, 95, 115, 123])

#Um viés constante para a função
bias = 1

#Define as classes
Y = np.array([1, 1, 1, -1, -1, -1])

#Stack compara os array e concatena 
X = np.vstack((peso, pH))

#quantidade de erros por registro
erro = np.zeros(len(Y))

#Função de ativação degau bipolar
def ativacao(valor):
  if valor < 0.0:
    return -1
  else:
    return 1

# learningRate / Lr, eta
eta = 0.1

#epocas / epoch
epocas = 10

#Cria uma matriz populando com 0, apenas para iniciar
W = np.zeros([1, 3])

#Percorres 
for j in range(epocas):
  print('*'+'-'*20)
  print('Epoca', j)
  for k in range(len(Y)):
    print('-'*20)
    #Insere/popula o vies do vetor X
    Xb = np.hstack((bias, X[:,k]))
    print('Vies, peso, pH: ',Xb)

    v = np.dot(W, Xb)

    # Calculando a saida do perceptron
    Yr = ativacao(v)
    print('Palpite (1 or -1): ',Yr)

    #Calculando o erro
    erro[k] = Y[k] - Yr
    print('Erros:', erro[k])

    #treinamento / backpropagation
    W = W + eta * erro[k] * Xb
    print('Treinamento:', W)
  print('Erros:', str(erro))
```
