# **Para alterar o tamanho da imagem pode ser utilizado o cv2.resize**

```python
!wget https://www.eurodicas.com.br/wp-content/uploads/2018/06/tudo-sobre-a-grecia-1.jpg

# Carregar Imagem
img = cv2.imread('grecia.jpg')

# Definindo a escala de 60%
escala = 60

# Definindo primeiro largura depois altura, shape 0 largura, shape 1 altura
width = int(img.shape[0] * escala / 100)
height = int(img.shape[1] * escala / 100)

# Mostra o valor de escala
print(width, height)

# Resize da imagem/ Tamanho para alterar (altura, largura) o tamanho da imagem/Interpolação: Forma como ele mescla os pixels
resized = cv2.resize(img, (height,width), interpolation = cv2.INTER_AREA)

# Printando a imagem
cv2_imshow(resized)

# Verificar tamanho
print(resized.shape)
```

# **Para rotacionar a imagem, pode ser utilizado o cv2.flip**

```python
# 0 flip vertical
imagem_vert = cv2.flip(resized, 0)
cv2_imshow(imagem_vert)
print('-'*20)

# 1 flip horizontal
imagem_hor = cv2.flip(resized, 1)
cv2_imshow(imagem_hor)
print('-'*20)

# -1 flip horizontal
imagem_invert = cv2.flip(resized, -1)
cv2_imshow(imagem_invert)
print('-'*20)
```

# **Convolução 2D (Filtrando imagens), pode ser usado cv2.filter2D()**

No processamento de imagem, um kernel, matriz de convolução ou máscara é uma pequena matriz usada para desfoque, nitidez, relevo, detecção de bordas e muito mais.

```python
import numpy as np

# Cria uma matriz de 5x5 de 0.04
kernel = np.ones((5,5), np.float32)/25
print(kernel)

# Faz a convolução da imagem deixando ela desfocada
conv = cv2.filter2D(resized, -2, kernel)
cv2_imshow(conv)
```

# **Convolução 2D (Filtrando imagens), pode ser usado cv2.blur**

Isso é feito convolvendo a imagem com um filtro de caixa normalizado. Ele simplesmen te
pega a média de todos os pixels sob a área do kernel e substitui o elemento central por es sa
média.

```python
imgProfile = cv2.imread('profile-example.jpg')
imgNoise = cv2.imread('noise_salt_and_pepper.png')

# cv2_imshow(imgProfile)
# cv2_imshow(imgNoise)
print('-'*20)

#Isso é feito convolvendo a imagem com um filtro de caixa normalizado. 
#Ele simplesmen te pega a média de todos os pixels sob a área do kernel e substitui o elemento central por es sa média.
blur = cv2.blur(imgProfile,(5,5))
cv2_imshow(blur)
print('-'*20)

#Aplica um filtro gaussiano, capaz de reduzir o nível de ruído de um sinal de entrada, a fim de diminuir a distorção numa imagem
blur_gau = cv2.GaussianBlur(imgProfile,(5,5),0)
cv2_imshow(blur_gau)
print('-'*20)

# Calcula a mediana de todos os pixels sob a janela do kernel e o pixel central é substituído por este valor mediano. 
# É altamente eficaz na remoção de ruídos de "sal e pimenta".
median = cv2.medianBlur(imgNoise,5)
cv2_imshow(median)
print('-'*20)


# É altamente eficaz na remoção de ruído preservando as bordas. Mas a operação é mais lenta em comparação com outros filtros.
blur_eficaz = cv2.bilateralFilter(imgProfile, 9, 75, 75)
cv2_imshow(blur_eficaz)
print('-'*20)
```

# **Draw rectangle**

A função cv2.rectangle() desenha um retângulo na imagem, se sua imagem for em tons de
cinza será necessário converter para RGB ou BGR para aparecer a cor das linhas
desenhadas.

Parâmetros:

**imagem:** É a imagem na qual o retângulo deve ser desenhado.

**start_point:** São as coordenadas iniciais do retângulo. As coordenadas são representadas como tuplas de dois valores, isto é ( valor da coordenada X , valor da coordenada Y ).

**end_point:** São as coordenadas finais do retângulo. As coordenadas são representadas como tuplas de dois valores, isto é ( valor da coordenada X , valor da coordenada Y ).

**cor:** é a cor da linha da borda do retângulo a ser desenhada. Para BGR , passamos uma tupla. por exemplo: (255, 0, 0) para a cor azul.

**espessura:** é a espessura da linha da borda do retângulo em px. Espessura de -1 px preencherá a forma do retângulo com a cor especificada.

```python
imgProfiles = cv2.imread('profile-example.jpg')
```

```python
# A função cv2.rectangle() desenha um retângulo na imagem
imgProfiles = cv2.imread('profile-example.jpg')

window_name = 'Image'
# Comece a coordenar, aqui (5, 5)
# representa o canto superior esquerdo do retângulo
start_point = (100, 75)
  
# Coordenada final, aqui (220, 220)
# representa o canto inferior direito do retângulo
end_point = (400, 475)
  
# Cor azul em BGR
color = (255, 0, 0)
  
# Espessura da linha de 2 px
thickness = 2

#Criando a imagem com o retangulo
imagem_qua = cv2.rectangle(imgProfiles, start_point, end_point,  color, thickness)
cv2_imshow(imagem_qua)
```

# **Template matching**

```python
img_rgb = cv2.imread('messi.png')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY) #Converte a imagem

#O 0 no final pega com escala de cinza para não precisar converter como acima
template = cv2.imread('messi_face.jpg', 0)

#Pega o tamanho da imagem
w, h = template.shape
print(w, h)

# Res cria uma matriz a partir do tamplate
res = cv2.matchTemplate(img_gray, template, cv2.TM_CCOEFF_NORMED)

# Localiza os valores mínimos e máximos (onde está a face)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
# print(min_val, max_val, min_loc, max_loc)

# Pega a localização do ponto, pega a largura e altura do template para fazer o ponto
bottom_right = (max_loc[0] + w, max_loc[ 1] + w)

#Cria o retangulo ao redor da face do messi, com a cor verde e expessura 2
cv2.rectangle(img_rgb, max_loc, bottom_right, ( 0, 255, 0), 2)

#Mostra a imagem
cv2_imshow(img_rgb)
```

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
#Carrega a imagem do mario 
img_rgb = cv2.imread('mario.png')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY) #altera para escala de cinza

#carrega o template, 0 abre a imagem em escala de cinza sem precisar de conversão
template = cv2.imread('coin.png',0)

#mostra o tamanho do template (w altura, h largura)
w, h = template.shape
# print(w, h)

# Res cria uma matriz a partir do tamplate
# matchtemplate faz uma busca do template na imagem, o cv2.TM_CC... é a forma como ele vai fazer a busca
res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)

# tamanho do limiar (semelhança) (a partir desse valor tem o objeto que busco)
threshold = 0.80

# Se tiver o threshold dentro de res
# somente valores maiores que o threshold
loc = np.where( res >= threshold)

#For nas localizações para fazer retangulo ao redor
#[::-1] inverte a localização
for pt in zip(*loc[::-1]):
    #primeiro imagem/ localização/
    cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)

#Mostra as moedas circularam
cv2_imshow(img_rgb)
```

# Abaixo, algoritimo utilizado para identificar uma circunferência, testado com moeda

Referências:
https://opencv24-python-tutorials.readthedocs.io/en/stable/py_tutorials/py_imgproc/py_houghcircles/py_houghcircles.html#hough-circles

https://theailearner.com/tag/cv2-houghcircles/

```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

#Carrega a moeda e muda a cor para cinza
img = cv2.imread('moedas.jpg',0)

#Reduz o ruído/suavisa com um pequeno Blur
img = cv2.medianBlur(img,5)

#Altera a imagem para escala de cinza
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)

#img=Carrega a imagem/ Método/ dp (razão inversa do acumulador e tamanho da imagem)/20= Distancia mínima do centro
#/param1=Limite superior/param2=limite do acumulador para os centros do círculo no estágio de detecção/minRadius=0,maxRadius=0 coloquei 0 pois é desconhecido
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,2,20,param1=150,param2=300,minRadius=0,maxRadius=0)

#cria o circulo 
circles = np.uint16(np.around(circles))

#Percorre a matriz e desenha o circulo
for i in circles[0,:]:
    # desenha o centro do círculo/cor vermelha
    cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
    # desenha o círculo externo/cor verde
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)

cv2_imshow(cimg)
#Mantem a imagem aberta até que seja teclado algo
cv2.waitKey(0)
#Fecha todos os frames
cv2.destroyAllWindows()
```
