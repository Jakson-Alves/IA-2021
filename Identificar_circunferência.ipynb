```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

#Carrega a moeda e muda a cor para cinza
img = cv2.imread('moedas.jpg',0)

#Reduz o ruído/suavisa com um pequeno Blur
img = cv2.medianBlur(img,5)

#Altera a imagem para escala de cinza
cimg = cv2.cvtColor(img,cv2.COLOR_GRAY2BGR)

#img=Carrega a imagem/ Método/ dp (razão inversa do acumulador e tamanho da imagem)/20= Distancia mínima do centro
#/param1=Limite superior/param2=limite do acumulador para os centros do círculo no estágio de detecção/minRadius=0,maxRadius=0 coloquei 0 pois é desconhecido
circles = cv2.HoughCircles(img,cv2.HOUGH_GRADIENT,2,20,param1=150,param2=300,minRadius=0,maxRadius=0)

#cria o circulo 
circles = np.uint16(np.around(circles))

#Percorre a matriz e desenha o circulo
for i in circles[0,:]:
    # desenha o centro do círculo/cor vermelha
    cv2.circle(cimg,(i[0],i[1]),2,(0,0,255),3)
    # desenha o círculo externo/cor verde
    cv2.circle(cimg,(i[0],i[1]),i[2],(0,255,0),2)

cv2_imshow(cimg)
#Mantem a imagem aberta até que seja teclado algo
cv2.waitKey(0)
#Fecha todos os frames
cv2.destroyAllWindows()
```
